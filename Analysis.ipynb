## Data Exploration

## Competitor Analysis, which coffee chains are the most prevalent in Los Angeles County


```python
%%sql
WITH cte AS (
SELECT 
	lb.name ,
	COUNT(*) AS count
FROM la_businesses lb 
GROUP BY lb.name 
ORDER BY COUNT(*) DESC
)
SELECT 
	name,
	count,
	DENSE_RANK() OVER(ORDER BY count DESC) AS ranking
FROM cte;
```

## Let us get a more comprehensive view on the top performing coffee shops

```python
%%sql
SELECT 
	lb.name ,
	COUNT(*) AS count,
	FORMAT(AVG(lb.rating),2) AS avg_rating,
    FORMAT(AVG(lb.price),2) AS avg_price,
	SUM(lb.review_count) AS total_reviews,
	COUNT(DISTINCT lb.zip_code) operating_zipcodes
FROM la_businesses lb 
GROUP BY lb.name 
ORDER BY COUNT(*) DESC;
```

## Here is a breakdown of each zip codes population, number of busiensses on yelp, their respective median income, and their population

```python
%%sql
SELECT 
	lb.zip_code ,
	cp.city_name ,
    cp.population,
	SUM(lb.review_count) AS total_review_count,
	FORMAT(AVG(lb.rating) , 2) AS avg_rating,
	COUNT(le.event_id) AS daily_events_within_area,
	lmi.median_income AS median_income
FROM la_businesses lb 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
JOIN la_events le 
	ON cp.zipcode = le.event_zip 
JOIN la_median_income lmi 
	ON le.event_zip = lmi.zipcode 
JOIN la_transactions lt 
	ON lb.id = lt.id 
GROUP BY lb.zip_code 
ORDER BY lb.zip_code;
```

## Let us have a ranking system for each numeric column

```python
%%sql
WITH cte AS (
SELECT 
	lb.zip_code ,
	cp.city_name ,
    FORMAT(AVG(lb.price),2) AS avg_price,
	SUM(lb.review_count) AS total_review_count,
	FORMAT(AVG(lb.rating) , 2) AS avg_rating,
	COUNT(le.event_id) AS daily_events_within_area,
	lmi.median_income AS median_income
FROM la_businesses lb 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
JOIN la_events le 
	ON cp.zipcode = le.event_zip 
JOIN la_median_income lmi 
	ON le.event_zip = lmi.zipcode 
JOIN la_transactions lt 
	ON lb.id = lt.id 
GROUP BY lb.zip_code 
)
SELECT 
	ROW_NUMBER() OVER() AS row_num,
	zip_code ,
	city_name,
	DENSE_RANK() OVER(ORDER BY total_review_count DESC) AS TOTAL_REVIEWS_RANKING,
	DENSE_RANK() OVER(ORDER BY avg_rating DESC) AS AVG_RATING_RANKING,
	DENSE_RANK() OVER(ORDER BY avg_price DESC) AS AVG_PRICE_RANKING,
	DENSE_RANK() OVER(ORDER BY daily_events_within_area DESC) AS EVENTS_RANKING,
	DENSE_RANK() OVER(ORDER BY median_income DESC) AS MEDIAN_INCOME_RANKING
FROM cte
GROUP BY zip_code
ORDER BY row_num;
```

## What is the average rating, median income?


```python
%%sql
SELECT 
	FORMAT(AVG(lb.rating),2) AS avg_rating,
	FORMAT(AVG(lmi.median_income),2) AS avg_income
FROM la_businesses lb 
JOIN la_median_income lmi 
	ON lb.zip_code = lmi.zipcode ;
```

##### Scenario 1: Luxury identity
- You are a prestigious coffee shop, and want to insert your business in a place with poor business ratings
- hoping to attract customers that are looking for luxury coffee experience
- rating 4.23 or below, need median income above $80,159.75

```python
%%sql
WITH cte AS (
SELECT 
	lb.zip_code ,
    cp.city_name,
    AVG(lb.price) AS avg_price,
	SUM(lb.review_count) AS total_review_count,
	FORMAT(AVG(lb.rating) , 2) AS avg_rating,
	COUNT(le.event_id) AS daily_events_within_area,
	lmi.median_income AS median_income
FROM la_businesses lb 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
JOIN la_events le 
	ON cp.zipcode = le.event_zip 
JOIN la_median_income lmi 
	ON le.event_zip = lmi.zipcode 
JOIN la_transactions lt 
	ON lb.id = lt.id 
GROUP BY lb.zip_code 
)
SELECT 
	zip_code,
    city_name,
    avg_rating,
    FORMAT(avg_price, 2) AS avg_price,
	CASE WHEN avg_rating <= 4.23 THEN "condition met" ELSE "no" END AS rating_condition,
    median_income,
	CASE WHEN median_income >= 80159.75 THEN "condition met" ELSE "no" END AS income_condition
FROM cte
ORDER BY rating_condition, income_condition;
```

## Scenario 2: Capturing market segment that allows business to take advantage of delivery

```python
%%sql
SELECT 
	lb.zip_code ,
	cp.city_name ,
	COUNT(lt.id) AS unavailable
FROM la_businesses lb 
JOIN la_transactions lt 
	ON lb.id = lt.id 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
WHERE lt.delivery = 'not_available'
GROUP BY 1
ORDER BY COUNT(lt.id)  DESC;
```

## zipcodes and their corresponding availability in delivery and no delivery (another way to visualize)


```python
%%sql
SELECT 
	lb.zip_code ,
	cp.city_name ,
	COUNT(CASE WHEN lt.delivery ='not_available' THEN lt.id ELSE NULL END) AS Count_of_deliver_unavailable,
	COUNT(CASE WHEN lt.delivery ='available' THEN lt.id ELSE NULL END) AS available
FROM la_businesses lb 
JOIN la_transactions lt 
	ON lb.id = lt.id 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
GROUP BY 1
ORDER BY Count_of_deliver_unavailable  DESC;
```

```python
%%sql
WITH cte AS (
SELECT 
	lb.zip_code ,
	cp.city_name ,
	COUNT(CASE WHEN lt.delivery ='not_available' THEN lt.id ELSE NULL END) AS Count_of_deliver_unavailable,
	COUNT(CASE WHEN lt.delivery ='available' THEN lt.id ELSE NULL END) AS available
FROM la_businesses lb 
JOIN la_transactions lt 
	ON lb.id = lt.id 
JOIN california_population cp 
	ON lb.zip_code = cp.zipcode 
GROUP BY 1
ORDER BY Count_of_deliver_unavailable  DESC
)
SELECT 
	zip_code,
	city_name,
	FORMAT(100*Count_of_deliver_unavailable / (Count_of_deliver_unavailable + available),2) AS percentage_unavailable
FROM cte;
```
